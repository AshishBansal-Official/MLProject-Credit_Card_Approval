```python
# EDA
```

```python
df['Age in days'] = abs(df['Age in days']/(30*12)).astype(int)
```

```python
df['Duration of work in days'] = abs(df['Duration of work in days']/(30*12)).round(2)
```

```python
df.rename(columns={'Age in days': 'Age', 'Duration of work in days': 'Duration of work'}, inplace=True)
```

```python
df['Status_paid_off_that_month'] = np.where(df['Status']=='C', 1, 0)
df['Status_No_loan_for_the_month'] = np.where(df['Status']=='X', 1, 0)
df['Status'] = df['Status'].replace({'C': -1, 'X': -2})
```

```python
df['Status'] = df['Status'].astype('int')
```

```python
df['Gender'].unique()
```

```python
df['Is there a car'].unique()
```

```python
df['Is there a property'].unique()
```

```python
df['Gender'] = np.where(df['Gender']=='M', 1, 0)
df['Is there a car'] = np.where(df['Is there a car']=='Y', 1, 0)
df['Is there a property'] = np.where(df['Is there a property']=='Y', 1, 0)
```

```python
df.head()
```

```python
df['Number of Children'].unique()
```

```python
map_children_col = {
    'No children': 0,
    '1 children': 1,
    '2+ children': 2,
}
```

```python
df['Number of Children'] = df['Number of Children'].map(map_children_col)
```

```python
df.head()
```

```python
df['Education Level'].unique()
```

```python
map_education_col = {
    'Lower secondary': 0,
    'Secondary / secondary special': 1,
    'Incomplete higher': 2,
    'Higher education': 3,
    'Academic degree': 4,
}
```

```python
df['Education Level'] = df['Education Level'].map(map_education_col)
```
